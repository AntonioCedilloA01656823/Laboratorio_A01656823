###¬†Carga de datos usando pandas.

```python
import pandas as pd #se importa pandas para poder leer el csv

df = pd.read_csv("covid19_tweets.csv")
df #Se verifica que la base de datos este correcta.
```

##¬†2. Analisis de variables con el uso de diversas graficas.

### 2.1 Cajas y bigotes. (Solo valores numericos)

```python
import seaborn as sb

#Creando cajas/bigotes de cada variable numerica con su respectivo titulo para diferenciar cual es cual
sb.boxplot(data = df["user_followers"],showfliers=False).set(title="User followers")


```

```python
sb.boxplot(data = df["user_friends"],showfliers=False).set(title="User friends")
```

```python
sb.boxplot(data = df["user_favourites"],showfliers=False).set(title="User favourites")
```

### 2.2 Histogramas

```python
import matplotlib.pyplot as plt #Importando librerias para la graficaci√≥n de histograma

fig, ax = plt.subplots()
ax.hist(df["user_followers"])
plt.title("User followers")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(df["user_friends"])
plt.title("User friends")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(df["user_favourites"])
plt.title("User favourites")
plt.show()
```
