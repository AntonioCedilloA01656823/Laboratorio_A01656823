```python
import pandas as pd

df = pd.read_csv("avocado.csv")
df
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

n = 3
kM = KMeans(n_clusters=n)

#arreglo de cluster
arr = []
for x in range(len(df["Total Bags"])):
    arr.append([df["Total Bags"][x],df["Small Bags"][x]])



fitArr = np.array(arr)
kM.fit(fitArr)
```

```python
centroides = kM.cluster_centers_
etiquetas = kM.labels_


plt.plot(fitArr[etiquetas==0,0],fitArr[etiquetas==0,1],'r.', label='cluster 1')
plt.plot(fitArr[etiquetas==1,0],fitArr[etiquetas==1,1],'b.', label='cluster 2')
plt.plot(fitArr[etiquetas==2,0],fitArr[etiquetas==2,1],'g.', label='cluster 3')

plt.plot(centroides[:,0],centroides[:,1],'mo',markersize=8, label='centroides')

plt.legend(loc='best')
plt.show()
```

```python
import math

def distancia(listaUno, listaDos):
    # dist = math.dist(listaUno, listaDos) #Igualmente se puede usar esta forma para la lista
    x1 = listaUno[0]
    y1 = listaUno[1]
    x2 = listaDos[0]
    y2 = listaDos[1]

    dist = math.sqrt(((x2 - x1)**2) - ((y2 - y1)**2))

    return dist
```
