## Obtenci√≥n de Mediana de los datos.

```python
#Forma 1 de conseguir la moda
mediana_followers = cDF['user_followers'].median()
mediana_friends = cDF['user_friends'].median()
mediana_favourites = cDF['user_favourites'].median()

mDF = pd.DataFrame({"Categoria":["User followers","User friends", "User favourites"],"Mediana":[mediana_followers,mediana_friends,mediana_favourites]})
mDF

```

```python
#Forma 2 de conseguir la moda

MDF = pd. DataFrame (cDF.median(numeric_only=True))
MDF
```

## Obtenci√≥n de la Moda de datos

```python
#Forma 2
 
cDF.mode()
```

## Obtenci√≥n de la media aritmetica 

```python
#Forma con pandas

promDF =  pd. DataFrame (cDF.mean(numeric_only=True))
promDF
```

## Obtenci√≥n de desviacion estandard.

```python
#Uso de pandas
stdDF = pd.DataFrame(cDF.std(numeric_only=True))
stdDF
```

```python
#uso de python/np

import numpy as np
fol_lis = cDF['user_followers'].values.tolist()
fr_lis =  cDF['user_friends'].values.tolist()
fav_lis = cDF['user_favourites'].values.tolist()
ver_lis = cDF['user_verified'].values.tolist()
rt_lis = cDF['is_retweet'].values.tolist()

fsd, frsd, favsd, versd, rtsd = np.std(fol_lis),np.std(fr_lis),np.std(fav_lis),np.std(ver_lis),np.std(rt_lis)

sdDF = pd.DataFrame({"Categoria":["User followers","User friends", "User favourites","User verified","Is retweet"],"Desv. estandar":[fsd,frsd,favsd,versd,rtsd]})
sdDF
```
